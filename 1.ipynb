```python
import numpy as np
from PIL import Image

# 1. Load the image
img = Image.open('butterfly.jpg')
img

# 2. Convert the image to a NumPy array
img_array = np.array(img)

# 3. Invert the pixels (This is the magic line!)
negative_array = 255 - img_array

# 4. Convert the array back into an image
negative_img = Image.fromarray(negative_array)

# 5. Save the result
negative_img.save('negative_image.jpg')
```

```python
plt.imshow(negative_img)
plt.axis("off")
plt.show()
```
