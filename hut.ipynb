```python
import matplotlib.pyplot as plt
```

```python
plt.plot(
    [2 , 6 , 14 , 14 , 6 , 2 , 2 , 6 , 6 , 4 , 2 , 4 , 12 , 14 ],
    [2 , 2 , 6 , 14 , 10 ,10 , 2 , 2 , 10 , 14 , 10 , 14 , 18 , 14] ,
    color = "brown"
)                    # structure

plt.plot(
    [3 , 3 , 5 , 5],
    [2 , 7 , 7 , 2],
    color = 'brown'
)                   # door

plt.plot(
    [ 8 , 8 , 12 , 12 , 8 ],
    [ 5 , 9 , 11 , 7  , 5 ],
    color = 'brown'
)                   # Window

plt.plot(
    [ 8 , 12 ],
    [ 7  , 9  ],
    color = 'brown'
)                   # window detail

plt.plot(
    [ 10 , 10 ],
    [ 6  , 10 ],
    color = 'brown'
)                   # window detail

plt.plot(
    [ 8 , 6 , 8 , 10 , 12 , 10],
    [ 11 , 15 , 16 , 12 , 13 , 17],
    color = 'brown'
)

plt.scatter(4 , 11.5 , s = 640 , color = 'red')

plt.axis("off")
plt.show()
```
