```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Set up Seaborn style
sns.set_theme(style="white")
plt.figure(figsize=(6, 6))

# 1. Generate geometry using angles (0 to 2*pi)
theta = np.linspace(0, 2 * np.pi, 300)

# Main Face Base (Perfect circle)
face_x = 3 + 2.5 * np.cos(theta)
face_y = 3.5 + 2.5 * np.sin(theta)

# Left & Right Eyes (Proper oval shapes like real emojis)
left_eye_x = 2.0 + 0.3 * np.cos(theta)
left_eye_y = 4.5 + 0.5 * np.sin(theta)

right_eye_x = 4.0 + 0.3 * np.cos(theta)
right_eye_y = 4.5 + 0.5 * np.sin(theta)

# Smile (Smooth mathematical arc at the bottom half of the circle)
smile_theta = np.linspace(np.pi + 0.3, 2 * np.pi - 0.3, 100)
smile_x = 3 + 1.4 * np.cos(smile_theta)
smile_y = 3.2 + 1.4 * np.sin(smile_theta)

# 2. Plotting the elements
# Yellow face fill with a slightly darker outline border
plt.fill(face_x, face_y, color="#FFCC00", edgecolor="#E6B800", linewidth=3)

# Fill the eyes
plt.fill(left_eye_x, left_eye_y, color="#333333")
plt.fill(right_eye_x, right_eye_y, color="#333333")

# Draw the smile with thick, rounded ends
sns.lineplot(
    x=smile_x, y=smile_y, color="#333333", linewidth=12, solid_capstyle="round"
)

# 3. Canvas constraints to keep it centered and borderless
plt.xlim(0, 6)
plt.ylim(0, 7)
plt.axis("off")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Circle
from matplotlib.patches import Ellipse

fig, ax= plt.subplots()

mycircle = Circle((4,4), 3.5,facecolor='yellow',edgecolor='black')
myellipse=Ellipse((2.5,5),width=1,height=1.5,facecolor='black',edgecolor='black')
my_ellipse=Ellipse((5.5,5),width=1,height=1.5,facecolor='black',edgecolor='black')

theta = np.linspace(np.pi/6, 5*np.pi/6, 100)

x = 4 + 1*np.cos(theta)
y = 3 - np.sin(theta)  

plt.plot(x, y, color='black', linewidth=2)

plt.axis('equal')
plt.xlim(0, 8)
plt.ylim(0, 8)

ax.add_patch(mycircle)
ax.add_patch(myellipse)
ax.add_patch(my_ellipse)

ax.set_aspect('equal')
ax.autoscale_view()   
ax.axis("off")
plt.show()
```
